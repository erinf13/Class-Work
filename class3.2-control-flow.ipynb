```python
num = input("Please enter a number between 1 and 10: ")
num = int(num)
if num < 3:
    print(f"num is less than 3: {num}")
elif num > 6:
    print(f"num is greater than 6: {num}")
else:
    print(f"num is between 3 and 6: {num}")
```

```python
# Let's make this better with try except 

num = input("Choose a number betwen 1 and 10")

try:
    num = int(num)
    if num < 3:
        print(f"num is less than 3: {num}")
    elif num > 6:
        print(f"num is greater than 6: {num}")
    else:
        print(f"num is between 3 and 6: {num}")
except Exception as e:
    print(f"Well that didn't work, try again :{e}")
```

```python
# type hinting, str, float, int, bool
# Nest if statements
variable: str = input("Please enter a value: ")

if variable.isnumeric():
    try:
        num = float = float(variable)
        if num > 10:
            print("the Value is too high, start over")
        else:
            print (f"{num} is a good choice")
    except Exception as e: 
        print(f"There was an error: {e}")
elif variable.isalpha():
    print(f"You chose, {variable.upper()}")
else:
    print("Please enter either a number or letter")
```

```python
number = input("Choose a number between 1 and 10!")

# and statement
if number.isnumeric():
    try:
        number = int(number)
        if number >= 1 and number <=10:
            print(f"You chose correctly: {number}")
        else:
            print(f"You suck! {number} is not between 1 and 10")
    except Exception as e:
        print(e)
    number = int(number)
else:
    print("Please choose a number")
```

```python
# or statement
# want a number not between 1 and 10 or -10 and -1

number = input("Please choose a number between -10 and -1 or 1 and 10: ")

try:
    if not number.isnumeric():
        raise Exception(f"{number} is not a number dummy, try reading!")
    number = int(number)
    if (number < -10 or number > -1) and (number < 1 or number > 10):
        raise Exception(f"{number} doesn't meet the rules, pay attention!")
    
    print(f"""
Hooray! You made it though: {number} meets the rules! 
Sorry if you chose a negative number, but '-' is not numeric!
""")
except Exception as e:
    print(f"There was an error: {e}")
```
