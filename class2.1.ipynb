```python
# string
string = "any sequence of characters" # strings use quotes "" or '' 
print(string)
print(type(string))

# never name variables this way - it should declare things such as the goal of the variable; s = bad variable name
s = str("something else")
print(type(s), s)

# use snake casting aka underscore to define variables, not hyphens, not NumberString or numberString
number_string = "123"

#Jupyter will print (display) the last line of a cell automatically
number_string
```

```python
# integer ; int
# whole numbers, positive or negative, without decimals

x = 1
y = int(3)

print(x)
print(type(y))

# aligning info to outputs
print("x plus y equals", x+y, ", but with commas")  # comma method aka no f string
print(f"x plus y equals {x + y}, but with an f string")  # f-string method

# concatenation is different from addition
string + str(x)  # converts integer x to string so it can be concatenated
```

```python
# float
# decimal numbers, positive or negative, with decimals

# we can overwrite variable names
x = 1.0
y = float(42)

print(f"x = {x}, y = {y}, and y is a {type(y)}")

# do math with float
#notice what happens to 5 
print(f"x + y = {x+y}, but y + 5 = {y+5}")
```

```python
# Boolean = true/false

is_true = True
is_false = bool(0)  # 0 is false, any other number is true
print(is_true)
print(is_false)


```

```python
# Variables store the resolution of code 
outcome = x + y
print(outcome)
```

# Methods
------------------------

```python
# string methods

print(type(string))
print(string.capitalize())  # capitalizes first letter
print(string.upper())       # makes all letters uppercase
print(string.isnumeric())  # checks if string is numeric
print(string.endswith("s"))  # checks if string ends with specified character(s)
```

```python
# floats and ints have few methods
x = 10
y = 5

# +, -, *, / are operators and % is modulus (remainder of division), ** is exponentiation
print(x*y)
print(x/y)
print(x%y, "modulus")
print(x/3, "division")
print(x//3, "floor division")
```
