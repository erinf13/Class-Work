## Num2Num Stats

- numerical to numerical statistics
- bivariate statistics

- correlation
    - correlation does NOT = causation
- need continuous numerical data
- quantify overlap between variables

### independent and dependent variables
- dependent variable
    - what you want to predict or explain
    - y 

- independent variable
    - the variable that explains or predicts the outcome of the dependent variable
    - x

### pearson coefficient output
- positive number = positive correlation
    - as the independent variable increases, the dependent variable also increases
- negative number means negative correlation
    - as the independent variable increases, the dependent variable decreases

### qualifications
- must be continuous data
    - not ranked data, ordinal data, or categorical data
- must have linear relationship
    - will not detect variability that expands or contracts at different points
- homoscedastic = equal variance
    - the variability of the dependent variable is the same at all levels of the independent variable


### ways to calculate
- scipy.stats.pearsonr(x, y)
    - includes p-value
- df.corr(numeric_only=True)
    - does not include p-value

```python
import pandas as pd
import scipy.stats as stats

url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"
df = pd.read_csv(url)
df.head(2)
```

```python
df.corr(numeric_only=True)
```

```python
for col1 in df.columns:
    for col2 in df.columns:
        iscol1num = pd.api.types.is_numeric_dtype(df[col1])
        iscol2num = pd.api.types.is_numeric_dtype(df[col2])
        if col1 == col2:
            continue
        if not iscol1num or not iscol2num:
            continue
        r, p = stats.pearsonr(df[col1], df[col2])
        print(f"{col1}-{col2} : r={r:.3f}, p={p:.3f}")
            
```

```python
r,p = stats.pearsonr(df["body_mass_g"], df["flipper_length_mm"])
print(r,p)
r2 = r**2
print(r2)
```

```python
r,p = stats.pearsonr(df["body_mass_g"], df["bill_length_mm"])
r2 = r**2

print(r,p,r2)
```
