```python
from functions.eda import eda
import pandas as pd 
url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"
df = pd.read_csv(url)
eda(df)
```

```python
# height_in_cm

from scipy import stats

heights = df["height_in_cm"]
sk = stats.skew(heights, nan_policy="omit")
kurt = stats.kurtosis(heights, nan_policy="omit")
print(f"skew = {sk:.3f}; kurt = {kurt:.3f}")
```

negative skewness = tail drawn out to left, outliers in the negative direction (more dispersed towards the left)

high kurtosis = a lot of the data is in the tail, longer and heavier tails

```python
# IQR for outliers
Q1 = heights.quantile(0.25)
Q3 = heights.quantile(0.75)

IQR = Q3 - Q1
Q1, Q3, IQR
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
hl = heights.loc[(heights<lower_bound) | (heights>upper_bound)].to_list()
hl
```

```python
def cm_to_ft(num):
    return (num * 0.393701) / 12
hl1 = [cm_to_ft(num) for num in hl]
hl1
```
