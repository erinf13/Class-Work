```python
count = 0
while True:
    if count >= 10:
        break #breaks the loop
    count += 1 # count = count + 1 
    print(count)
```

```python
# A better option
count = 0

while count < 10:
    count += 1
    print(count)
```

```python
is_pass_complete: bool = False

while not is_pass_complete:
    password = input("Enter password in alphabetic characters")
    if not password.isalpha():
        print("Please choose another password with only alphabetic characters")
        continue # continues the loop from here, skipping the rest below
    is_pass_complete = True
    print("Password set")
```

```python
from typing import List
# for loop

nums: List[int] = [1, 2, 3, 4, 5]

for num in nums:
# num =  temporary variable used to call back each 
# item in the collection in the code block
    num += 1
    if num % 2 != 0: # if odd
        #break would break the loop
        continue # do not print anything odd
    print(num)
```

```python
sentence: str = "This is a Python course."

words: List[str] = sentence.split(" ")
used_vowels = []

for word in words:
    vowels = ["a", "e", "i", "o", "u","y"]
    for letter in word:
        if letter in vowels:
            used_vowels.append(letter)
print(used_vowels)
```

```python
# what are functions
# function take an input (argument), processes it, and return
# an output (return)

def shorten_names(names_list: List[str], chars: int = 3) -> List[str]:
    """
    Returns a list of strings that are the first three
    characters of the word with a period appended.
    
    Args:
        names_list (List[str]): A List of strings that will be
        shortened.
        chars (int): The number of characters to shorten each
        string to. Must be greater than 0. Defaults to 3
    Return:
        List[str]: A list of shortened strings to 3 characters.
    """
    if type(chars) != int:
        raise TypeError("Chars must be an int")
    if chars < 0:
        raise ValueError("Chars must be greater than 0")
    names_list = [f"{name[:chars]}." for name in names_list]
    return names_list
```

```python
bar_names: List[str] = ["Ottoman Empire", "France", "Spain"]

shorten_names(bar_names, chars=2)
```
