## Simple Variables 

- string (str)
- int
- float
- boolean

```python
# string
string = "any sequence of characters" # strings use quotes "" or '' 
print(string)
print(type(string))

# never name variables this way - it should declare things such as the goal of the variable; s = bad variable name
s = str("something else")
print(type(s), s)

# use snake casting aka underscore to define variables, not hyphens, not NumberString or numberString
number_string = "123"

#Jupyter will print (display) the last line of a cell automatically
number_string
```

```python
# string methods

print(type(string))
print(string.capitalize())  # capitalizes first letter
print(string.upper())       # makes all letters uppercase
print(string.isnumeric())  # checks if string is numeric
print(string.endswith("s"))  # checks if string ends with specified character(s)
```

```python
# integer ; int
# whole numbers, positive or negative, without decimals

x = 1
y = int(3)

print(x)
print(type(y))

# aligning info to outputs
print("x plus y equals", x+y, ", but with commas")  # comma method aka no f string
print(f"x plus y equals {x + y}, but with an f string")  # f-string method

# concatenation is different from addition
string + str(x)  # converts integer x to string so it can be concatenated
```

```python
# float
# decimal numbers, positive or negative, with decimals

# we can overwrite variable names
x = 1.0
y = float(42)

print(f"x = {x}, y = {y}, and y is a {type(y)}")

# do math with float
#notice what happens to 5 
print(f"x + y = {x+y}, but y + 5 = {y+5}")
```

```python
# floats and ints have few methods
x = 10
y = 5

# +, -, *, / are operators and % is modulus (remainder of division), ** is exponentiation
print(x*y)
print(x/y)
print(x%y, "modulus")
print(x/3, "division")
print(x//3, "floor division")
```

```python
# Boolean = true/false

is_true = True
is_false = bool(0)  # 0 is false, any other number is true
print(is_true)
print(is_false)


```

```python
# Variables store the resolution of code 
outcome = x + y
print(outcome)
```

## Complex Variables (collections)

- set
- tuple
- list
- dictionary
- combine collections

```python
# set
# no duplicates, no indexed (unordered), no static order, is iterable (can loop through it - do something repeatidly), 
#   is a collection, is mutable (can be changed)

set1 = {"Ben", "Jason", "Darlene", "Jason", "Erin"}
set_declared = set(["Ben", "Jason", "G", "G", "Darlene"])

print(f"set1 = {set1}")
print(f"set_declared = {set_declared}")
```

```python
# Method to find overlaps
overlaps = set1.intersection(set_declared)
overlaps 
```

```python
# Method to find difference between two sets
diff = set1.difference(set_declared)
diff
```

```python
# Method to add to a set
set1.add("Halle")
set1
```

```python
# Method to remove from a set
set1.remove("Darlene")
set1
```

```python
# Tuple
# non-mutable (cannot be changed), ordered (indexed), allows duplicates, is iterable, is a collection
tuple1 = (1,2,3,4)
tuple_declared = tuple((3,4,5,6))

print(f"tuple1 = {tuple1}")
print(f"tuple_declared = {tuple_declared}")
```

```python
tuple1 = (1,2,3,4,5) # overwriting, not changing
tuple1
```

```python
tuple1.count(2)  # counts how many times 2 appears in the tuple
```

```python
tuple1.index(1) # return index of value (position of first occurrence)
```

```python
print(f"tuple1 = {tuple1}")
tuple1[0] # return value at index 0
```

```python
# list
# mutable (can be changed), ordered (indexed), allows duplicates, is iterable, is a collection
list1 = ["G","G", "Paola", "Emil"]
list_declared = list(("G", "G", "Paola", "Emil"))

print(f"list1 = {list1}")
print(f"list_declared = {list_declared}")
```

```python
# Index Notation [:]
print(f"list1 = {list1}")
print(f"first item in list1 = {list1[0]}")
print(f"last item in list1 = {list1[-1]}")
print(f"series between 2 & 3 items = {list1[1:3]}") # second item is exclusive
print(f"everything before index 3 = {list1[:3]}")
print(f"everything after the first item = {list1[1:]}") 
```

```python
# list methods

# add 
list1.append("Clare")
list2 = list1.append("Nathan")
print(f"appended Clare = {list1}")
print(f"list2 = {list2}")  # list2 is None because append does not return anything

# remove
removed = list1.pop()
print(f"removed {removed}. New list1 is {list1}")

# change index
list1[1] = "J"
print(f"Index 1 of list1 changed. New list1 = {list1}")

# combine lists
combined_list1 = [list1, list_declared] # creates a list of lists (two-dimensional list)
print(f"combined_list1 = {combined_list1}")

combined_lists = [*list1, *list_declared] # unpacks both lists into a new list (creates a new place in memory)
print(f"combining lists into one = {combined_lists}")

list1.extend(list_declared) # extends list1 by adding elements from list_declared to the end of list1, modifies list1 in place
print(f"list1 after extension = {list1}") 
```

```python
# dictionary
# key:value pairs, indexed by key, not ordered, duplicates allowed for values but not keys, is mutable (can be changed)
# {key: value, key1: value}
dict1 = {"name": "Erin", "age": 21}
dict_declared = dict(name="Erin", age=21) # we never do this

print(dict_declared)
dict1
```

```python
# get values
dict1["name"]
```

```python
# change value
dict1["name"] = "Erin F"
dict1
```

```python
# call keys
dict1.keys()
```

```python
# call values
dict1.values()
```
