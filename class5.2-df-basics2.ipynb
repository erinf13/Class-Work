```python
# itertuples () is faster

for item in df.itertuples():
    print(item)
```

```python
for item in df.itertuples():
    print(item.name) # named tuples use dot notation 
```

```python
for idx, id, n, p, y, lat, long, _ in df.itertuples():
    print(p) 
```

```python
for item in df.itertuples():
    if item.place == "Denmark":
        df.at[item.Index, "place"] = "Den."
        
df["place"].unique() # get unique values in a column
```

```python
# Englad => Eng.
# if only changing one name, use .loc[]

df.loc[df["place"] == "England", "place"] = "Eng."
df["place"].unique()
```

```python
# .replace() for multiple

# hash = {
#     "Holy Roman Empire": "HRE",
#     "Geneva": "Gen.",   
# }

hash = {name:f"{name[:3]}." for name in df["place"].unique() if len(name) != 4}
hash

```

```python
df["place"] = df["place"].replace(hash)
df["place"].unique()
```

```python
# pandas operators & (and), | (or), ~ (not)

# find ambassador at location at same time

df.loc[df["place"] == "Den."].count()
```

```python
df.loc[(df["place"] == "Den.") & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
df.loc[(df["place"] == "Den.") & (df["year"] < 1561) & (df["year"] > 1550)]
```

```python
# use ~ for not 
# how many ambassadors NOT in Den. between 1550 and 1561
df.loc[~(df["place"] == "Den.") & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
important_ambs = ["Eng.", "Ott.", "Ven."]

df.loc[(df["place"].isin(important_ambs)) & (df["year"] < 1561) & (df["year"] > 1550)].count()
```

```python
for item in df["place"].unique():
    if item in important_ambs:
        print(f"is {item} in important ambs list? {item in important_ambs} ")
```
