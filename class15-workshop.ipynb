```python
import pandas as pd
import scipy.stats as stats

url = "https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/penguins_cleaned.csv"
df = pd.read_csv(url)
df.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# on a plot , never leave out the scatter, but sometimes leaving out the line of best fit can be helpful to see the data more clearly

plt.figure(figsize=(5,3))

r,p = stats.pearsonr(df["bill_length_mm"], df["bill_depth_mm"])
r2 = r**2

text = f"""r={r:.4f}
p={p:.4f}
r2={r2:.4f}"""

reg = sns.regplot(
    data=df,
    x = "bill_length_mm",
    y = "bill_depth_mm",
    line_kws={"color":"red", "label" : "Line of best fit"})
reg.text(
    x=0.1,
    y=0.1, 
    s=text,
    transform=reg.transAxes, 
    bbox={"boxstyle":"round, pad = 0.5", "fc":"white"})
reg.set(
    title= "Figure 1. Correlation between Bill Length and Bill Depth",
    xlabel="Bill Length (mm)",
    ylabel="Bill Depth (mm)"
)
plt.legend()
plt.show()
```
